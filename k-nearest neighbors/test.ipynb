```python
from sklearn.neighbors import KNeighborsClassifier
import nearestK # K-nearest neighbors

import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
cmap = ListedColormap(['#FF0000','#00FF00','#0000FF']) # Định dạng màu sắc theo mã Hex

iris = datasets.load_iris() # Tập dữ liệu các loài hoa iris
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

plt.figure()
plt.scatter(X[:,2],X[:,3], c=y, cmap=cmap, edgecolor='k', s=20)
plt.show()
```

## Scikit-learn's KNN

```python
knn = KNeighborsClassifier(n_neighbors=5, p=2)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print("Các dự đoán từ KNN của scikit-learn:",pred)

acck = np.sum(pred == y_test) / len(y_test)
print(f"Độ chính xác: {acck*100}%")
```

```python
plt.figure()
plt.scatter(X_train[:,2],X_train[:,3], c=y_train, cmap=cmap, edgecolor='k', s=20)
plt.scatter(X_test[:,2], X_test[:,3], c=pred, cmap=cmap)
plt.show()
```

## nearestK's KNN

```python
nnk = nearestK.KNearestClassifier(k=5, p=2)
nnk.fit(X_train, y_train)
predictions = nnk.predict(X_test)

print("Các dự đoán từ KNN của nearestK.py:",predictions)

acc = np.sum(predictions == y_test) / len(y_test)
print(f"Độ chính xác: {acck*100}%")
```

```python
plt.figure()
plt.scatter(X_train[:,2],X_train[:,3], c=y_train, cmap=cmap, edgecolor='k', s=20)
plt.scatter(X_test[:,2], X_test[:,3], c=predictions, cmap=cmap)
plt.show()
```
