```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import LR # Linear Regression
```

```python
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=4)
```

```python
# LinearRegression từ scikit-learn
reg = LinearRegression()
reg.fit(X, y)
pred = reg.predict(X)
# LinearRegression từ LR
reg_scratch = LR.LinearRegression()
reg_scratch.fit(X, y)
pred_scratch = reg_scratch.predict(X)
```

```python
# Tính sai số bình phương trunh bình
def MSE(predictions, y):
    return ((1/len(predictions))* np.sum(np.square(y-predictions)))

print("Sai số bình phương trung bình cho scikit-learn's LinearRegression:",MSE(pred, y))
print("Sai số bình phương trunh bình của LR:", mean_squared_error(y, pred_scratch))
```

```python
plt.scatter(X, y, color = 'b')
```

```python
# scikit-learn's Linear Regression
plt.scatter(X, y, color = 'b')
plt.plot(X, pred, color = 'red', label = 'scikit-learn\'s LinearRegression' )
plt.legend()
plt.show()
```

```python
# LR 
plt.scatter(X, y, color = 'b')
plt.plot(X, pred_scratch, color = 'green', label = 'LR')
plt.legend()
plt.show()
```
